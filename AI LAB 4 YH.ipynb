```python
# TASK 1
import pandas as pd

# Dataset: Weather, Temperature, Play (Yes or No)
data = {
    'Weather': ['Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Rainy', 'Overcast', 'Sunny', 'Sunny', 'Rainy', 'Sunny', 'Overcast', 'Overcast', 'Rainy'],
    'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'],
    'Play': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'Yes', 'Yes', 'Yes', 'No']
}

# Convert to a pandas DataFrame
df = pd.DataFrame(data)

# Step 1: Calculate prior probabilities for Play = Yes and Play = No
prior_play_yes = df['Play'].value_counts(normalize=True)['Yes']
prior_play_no = df['Play'].value_counts(normalize=True)['No']

# Step 2: Calculate the likelihoods P(Feature | Play=Yes) and P(Feature | Play=No)
likelihood_play_yes = df[df['Play'] == 'Yes'].apply(lambda x: x.value_counts(normalize=True)).T
likelihood_play_no = df[df['Play'] == 'No'].apply(lambda x: x.value_counts(normalize=True)).T

# Step 3: Given input features 'Overcast' (Weather) and 'Mild' (Temperature)
weather_query = 'Overcast'
temperature_query = 'Mild'

# Calculate the likelihoods for the query features given Play=Yes and Play=No
likelihood_weather_given_yes = likelihood_play_yes.loc['Weather', weather_query] if weather_query in likelihood_play_yes.loc['Weather'] else 0
likelihood_temp_given_yes = likelihood_play_yes.loc['Temperature', temperature_query] if temperature_query in likelihood_play_yes.loc['Temperature'] else 0

likelihood_weather_given_no = likelihood_play_no.loc['Weather', weather_query] if weather_query in likelihood_play_no.loc['Weather'] else 0
likelihood_temp_given_no = likelihood_play_no.loc['Temperature', temperature_query] if temperature_query in likelihood_play_no.loc['Temperature'] else 0

# Step 4: Apply Bayes' Theorem to calculate the posterior probabilities
p_yes_given_features = prior_play_yes * likelihood_weather_given_yes * likelihood_temp_given_yes
p_no_given_features = prior_play_no * likelihood_weather_given_no * likelihood_temp_given_no

# Output the results
print(f"P(Play=Yes | Weather={weather_query}, Temperature={temperature_query}) = {p_yes_given_features}")
print(f"P(Play=No | Weather={weather_query}, Temperature={temperature_query}) = {p_no_given_features}")

# Step 5: Make the prediction based on the higher probability
if p_yes_given_features > p_no_given_features:
    print("\nPrediction: The players can play (Play=Yes).")
else:
    print("\nPrediction: The players cannot play (Play=No).")
```

```python

# TASK 2 
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Step 1: Define the dataset
data = {
    'age': ['youth', 'youth', 'middle_aged', 'senior', 'senior', 'senior', 'middle_aged', 'youth', 'youth', 'senior', 'youth', 'middle_aged', 'middle_aged', 'senior'],
    'income': ['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'medium', 'low', 'medium', 'medium', 'medium', 'high', 'medium'],
    'student': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no'],
    'credit_rating': ['fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'excellent'],
    'buys_computer': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
}

# Step 2: Convert the dataset into a pandas DataFrame
df = pd.DataFrame(data)

# Step 3: Initialize separate LabelEncoders for each categorical column
label_encoders = {
    'age': LabelEncoder(),
    'income': LabelEncoder(),
    'student': LabelEncoder(),
    'credit_rating': LabelEncoder(),
    'buys_computer': LabelEncoder()
}

# Step 4: Encode each feature column and the target column
df['age_encoded'] = label_encoders['age'].fit_transform(df['age'])
df['income_encoded'] = label_encoders['income'].fit_transform(df['income'])
df['student_encoded'] = label_encoders['student'].fit_transform(df['student'])
df['credit_rating_encoded'] = label_encoders['credit_rating'].fit_transform(df['credit_rating'])
df['buys_computer_encoded'] = label_encoders['buys_computer'].fit_transform(df['buys_computer'])

# Step 5: Display the encoded DataFrame
print("Encoded DataFrame:")
print(df[['age', 'income', 'student', 'credit_rating', 'buys_computer', 
          'age_encoded', 'income_encoded', 'student_encoded', 'credit_rating_encoded', 'buys_computer_encoded']])

# Step 6: Calculate the prior probabilities (P(Yes) and P(No))
prior_yes = df['buys_computer_encoded'].value_counts(normalize=True)[1]  # P(Yes)
prior_no = df['buys_computer_encoded'].value_counts(normalize=True)[0]   # P(No)

# Step 7: Calculate the likelihoods for each feature given the class
# P(age|Yes), P(income|Yes), P(student|Yes), P(credit_rating|Yes)
likelihood_age_given_yes = df[df['buys_computer_encoded'] == 1]['age_encoded'].value_counts(normalize=True)
likelihood_income_given_yes = df[df['buys_computer_encoded'] == 1]['income_encoded'].value_counts(normalize=True)
likelihood_student_given_yes = df[df['buys_computer_encoded'] == 1]['student_encoded'].value_counts(normalize=True)
likelihood_credit_rating_given_yes = df[df['buys_computer_encoded'] == 1]['credit_rating_encoded'].value_counts(normalize=True)

# P(age|No), P(income|No), P(student|No), P(credit_rating|No)
likelihood_age_given_no = df[df['buys_computer_encoded'] == 0]['age_encoded'].value_counts(normalize=True)
likelihood_income_given_no = df[df['buys_computer_encoded'] == 0]['income_encoded'].value_counts(normalize=True)
likelihood_student_given_no = df[df['buys_computer_encoded'] == 0]['student_encoded'].value_counts(normalize=True)
likelihood_credit_rating_given_no = df[df['buys_computer_encoded'] == 0]['credit_rating_encoded'].value_counts(normalize=True)

# Step 8: Given the query (youth/medium/yes/fair), encode the query values using separate LabelEncoders
query = ['youth', 'medium', 'yes', 'fair']

# Use the label encoders to transform the query values, ensuring that it is consistent with the training data
encoded_query = [
    label_encoders['age'].transform([query[0]])[0],  # 'youth'
    label_encoders['income'].transform([query[1]])[0],  # 'medium'
    label_encoders['student'].transform([query[2]])[0],  # 'yes'
    label_encoders['credit_rating'].transform([query[3]])[0]   # 'fair'
]

# Step 9: Apply Bayes' Theorem to calculate the posterior probabilities

# P(Yes | query) = P(query | Yes) * P(Yes)
p_yes_given_query = prior_yes
p_yes_given_query *= likelihood_age_given_yes.get(encoded_query[0], 0)  # P(age|Yes)
p_yes_given_query *= likelihood_income_given_yes.get(encoded_query[1], 0)  # P(income|Yes)
p_yes_given_query *= likelihood_student_given_yes.get(encoded_query[2], 0)  # P(student|Yes)
p_yes_given_query *= likelihood_credit_rating_given_yes.get(encoded_query[3], 0)  # P(credit_rating|Yes)

# P(No | query) = P(query | No) * P(No)
p_no_given_query = prior_no
p_no_given_query *= likelihood_age_given_no.get(encoded_query[0], 0)  # P(age|No)
p_no_given_query *= likelihood_income_given_no.get(encoded_query[1], 0)  # P(income|No)
p_no_given_query *= likelihood_student_given_no.get(encoded_query[2], 0)  # P(student|No)
p_no_given_query *= likelihood_credit_rating_given_no.get(encoded_query[3], 0)  # P(credit_rating|No)

# Step 10: Make the prediction based on the higher probability
print(f"P(Yes | query) = {p_yes_given_query}")
print(f"P(No | query) = {p_no_given_query}")

if p_yes_given_query > p_no_given_query:
    print("\nPrediction: The person will buy a computer (Class = Yes).")
else:
    print("\nPrediction: The person will not buy a computer (Class = No).")
```

```python
#TASK NO.2 YHINA

import pandas as pd
import numpy as np

# Step 1: Define the dataset
data = {
    'age': ['youth', 'youth', 'middle_aged', 'senior', 'senior', 'senior', 'middle_aged', 'youth', 'youth', 'senior', 'youth', 'middle_aged', 'middle_aged', 'senior'],
    'income': ['high', 'high', 'high', 'medium', 'low', 'low', 'low', 'medium', 'low', 'medium', 'medium', 'medium', 'high', 'medium'],
    'student': ['no', 'no', 'no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no'],
    'credit_rating': ['fair', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'fair', 'fair', 'excellent', 'excellent', 'fair', 'excellent'],
    'buys_computer': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Step 2: Encode categorical variables manually
# Define encoding for categorical variables
encoding = {
    'age': {'youth': 0, 'middle_aged': 1, 'senior': 2},
    'income': {'high': 0, 'medium': 1, 'low': 2},
    'student': {'no': 0, 'yes': 1},
    'credit_rating': {'fair': 0, 'excellent': 1},
    'buys_computer': {'no': 0, 'yes': 1}
}

# Apply encoding
df['age_encoded'] = df['age'].map(encoding['age'])
df['income_encoded'] = df['income'].map(encoding['income'])
df['student_encoded'] = df['student'].map(encoding['student'])
df['credit_rating_encoded'] = df['credit_rating'].map(encoding['credit_rating'])
df['buys_computer_encoded'] = df['buys_computer'].map(encoding['buys_computer'])

# Display encoded dataframe
print("Encoded DataFrame:")
print(df[['age', 'income', 'student', 'credit_rating', 'buys_computer', 'age_encoded', 'income_encoded', 'student_encoded', 'credit_rating_encoded', 'buys_computer_encoded']])

# Step 3: Calculate Prior Probabilities (P(yes) and P(no))
total_instances = len(df)
p_yes = len(df[df['buys_computer_encoded'] == 1]) / total_instances
p_no = len(df[df['buys_computer_encoded'] == 0]) / total_instances

# Step 4: Calculate Likelihoods for each feature given the class
# Function to calculate P(feature | class)
def calculate_likelihood(df, feature, class_label):
    feature_class_data = df[df['buys_computer_encoded'] == class_label]
    feature_counts = feature_class_data[feature].value_counts(normalize=True)
    return feature_counts

# Likelihoods for 'yes' and 'no' classes
likelihoods_yes = {}
likelihoods_no = {}

# List of features to compute likelihoods for
features = ['age_encoded', 'income_encoded', 'student_encoded', 'credit_rating_encoded']

for feature in features:
    likelihoods_yes[feature] = calculate_likelihood(df, feature, 1)  # P(feature | yes)
    likelihoods_no[feature] = calculate_likelihood(df, feature, 0)   # P(feature | no)

# Display the likelihoods
print("\nLikelihoods for 'yes' class:")
for feature, likelihood in likelihoods_yes.items():
    print(f"P({feature} | yes): {likelihood}")

print("\nLikelihoods for 'no' class:")
for feature, likelihood in likelihoods_no.items():
    print(f"P({feature} | no): {likelihood}")

# Step 5: Query to classify (youth, medium, yes, fair)
query = ['youth', 'medium', 'yes', 'fair']
encoded_query = [
    encoding['age'][query[0]],         # 'youth'
    encoding['income'][query[1]],      # 'medium'
    encoding['student'][query[2]],     # 'yes'
    encoding['credit_rating'][query[3]]  # 'fair'
]

print("\nEncoded query:", encoded_query)

# Step 6: Calculate Posterior Probabilities using Bayes' Theorem
# P(class | features) = P(features | class) * P(class) / P(features)

# P(yes | query)
p_yes_given_query = p_yes
for feature, value in zip(features, encoded_query):
    p_yes_given_query *= likelihoods_yes.get(feature, {}).get(value, 0)

# P(no | query)
p_no_given_query = p_no
for feature, value in zip(features, encoded_query):
    p_no_given_query *= likelihoods_no.get(feature, {}).get(value, 0)

# Display the posterior probabilities
print("\nPosterior probabilities:")
print(f"P(yes | query) = {p_yes_given_query}")
print(f"P(no | query) = {p_no_given_query}")

# Step 7: Make the classification decision
if p_yes_given_query > p_no_given_query:
    print("\nPrediction: The player will buy a computer (Class = 'yes').")
else:
    print("\nPrediction: The player will not buy a computer (Class = 'no').")
```
