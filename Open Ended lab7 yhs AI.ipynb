```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score, silhouette_score
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# Load dataset (you can replace this with your own dataset)
data = load_iris()
X = data.data
y = data.target

# Split data into training and testing sets for supervised models
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize supervised models
models = {
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier()
}

# Dictionary to store accuracy of each supervised model
accuracies = {}

# Train and test each supervised model, then store accuracy
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracies[name] = accuracy

# For KMeans, we use the silhouette score (unsupervised metric)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)  # Use the full dataset for clustering
sil_score = silhouette_score(X, kmeans.labels_)
accuracies["KMeans"] = sil_score  # Store silhouette score for KMeans

# Create a bar chart comparing the accuracies
plt.figure(figsize=(10, 6))
plt.bar(accuracies.keys(), accuracies.values(), color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Model')
plt.ylabel('Score')
plt.title('Model Comparison (Supervised and Unsupervised)')
plt.ylim(0, 1)  # For KNN, Naive Bayes, Decision Tree
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# Load the Iris dataset
iris = load_iris()
data = iris.data  # Features (X)
y_true = iris.target  # True labels (y)

# Encode the true labels (if necessary) to match the cluster labels
encoder = LabelEncoder()
y_true_encoded = encoder.fit_transform(y_true)

# Placeholder for accuracies
k_values = range(2, 11)  # Vary the number of clusters from 2 to 10
accuracies = []

for k in k_values:
    # Initialize and fit K-Means
    kmeans = KMeans(n_clusters=k, random_state=42)
    y_kmeans = kmeans.fit_predict(data)
    
    # Map predicted clusters to true labels
    mapping = {}
    for i in range(k):
        cluster_labels = y_true_encoded[y_kmeans == i]
        if len(cluster_labels) > 0:
            mapping[i] = np.bincount(cluster_labels).argmax()
        else:
            mapping[i] = -1  # Handle empty clusters
    
    # Map predicted clusters to true labels
    y_pred = np.array([mapping[label] for label in y_kmeans])
    
    # Calculate accuracy
    accuracy = accuracy_score(y_true_encoded, y_pred)
    accuracies.append(accuracy)

# Plot the accuracy graph
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o', linestyle='--', color='b', label='K-Means Accuracy')
plt.title('K-Means Clustering Accuracy vs. Number of Clusters (k)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.legend()
plt.grid()
plt.show()
```
